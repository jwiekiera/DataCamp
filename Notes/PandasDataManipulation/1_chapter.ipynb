## 1. Transforming DataFrames

```python
print(zoo_df.values)
print(zoo_df.columns)
print(zoo_df.index)
```

### 1.2: Sorting and subsetting
#### Sorting values

```python
# Sorting DataFrame by values of column eggs ascending
zoo_df.sort_values("legs").head()
```

```python
# Sorting DataFrame by values of column eggs descending
zoo_df.sort_values("legs", ascending=False).head()
```

```python
# Sorting DataFrame by values of columns eggs and tail (passing list)
zoo_df.sort_values(["eggs","tail"]).head()
```

```python
# Sorting DataFrame by values of columns eggs and tail with list of sorting order
zoo_df.sort_values(["eggs","tail"], ascending=[True,False]).head()
```

#### Subsetting 
Subsetting columns

```python
print(zoo_df["eggs"].head())
print(zoo_df[["eggs","tail"]].head())
```

Subsetting rows

```python
# Printing sum of rows with eggs equal to 1 (True)
(zoo_df["eggs"] == 1).sum()
```

```python
# Printing rows with eggs equal to 1 (True)
zoo_df[zoo_df["eggs"] == 1].head()
```

```python
zoo_df[zoo_df["animal_name"] == "worm"]
```

Subsetting based on multiple conditions

```python
# Printing values with animals that have hair and give milk - first way
gives_milk = zoo_df["milk"] == 1
have_hair = zoo_df["hair"] == 1
zoo_df[gives_milk & have_hair].head()
```

```python
# Printing values with animals that have hair and give milk - second way
zoo_df[(zoo_df["milk"] == 1) & (zoo_df["hair"] == 1)].head()
```

```python
# Subsetting using .isin()
is_worm_or_girl = zoo_df["animal_name"].isin(["worm","girl"])
zoo_df[is_worm_or_girl]
```

### 1.3: New columns

```python
 # Creating new column based on milk and hair column
zoo_df['milk_and_hair'] = (zoo_df['milk'] == 1) & (zoo_df['hair'] == 1)
zoo_df[['animal_name','milk','hair','milk_and_hair']].head()
```
